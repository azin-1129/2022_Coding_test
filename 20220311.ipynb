```python
#8-1

def fibo(x):
    print("fibo",x,"호출중")
    if x==1 or x==2:
        return 1
    return fibo(x-1)+fibo(x-2)

print(fibo(4))
```

```python
#8-2

d=[0]*100

def fibo(x):
    print("fibo",x,"호출중")
    if x==1 or x==2:
        return 1
    if d[x]!=0:
        return d[x]
    
    d[x]=fibo(x-1)+fibo(x-2)
    
    return d[x]

print(fibo(99))
```

```python
#2. 217p

X=int(input())

count=0

while (X>=2):
    print("X:",X)
    if X%5==0:
        X//=5
    elif X%3==0:
        X//=3
    elif X%2==0:
        X//=2
    else:
        X-=1
    count+=1
    
print(count)
```
