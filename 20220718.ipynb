### 2. 교보문고 베스트셀러 책이름, 저자, 가격 출력

```python
html=urlopen('http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf')
obj=BeautifulSoup(html, 'html.parser')
```

```python
obj
```

```python
book_page_urls=[]

for cover in obj.find_all('div',{'class':'detail'}): # 각 책
    link=cover.select('a')[0].get('href') # 책 이름란의 href
    book_page_urls.append(link)
```

```python
for index, book_page_url in enumerate(book_page_urls):
    html=urlopen(book_page_url)
    obj=BeautifulSoup(html,'html.parser')
    
    title=obj.find('meta',{'property':'og:title'}).get('content')
    author=obj.select('span.name a')[0].text
    image=obj.find('meta',{'property':'og:image'}).get('content')
    url=obj.find('meta',{'property':'og:url'}).get('content')
    Price=obj.find('meta',{'property':'og:price'}).get('content')
    
    print(index+1,title,author,image,url,Price)
```

### 3. 네이버 베스트셀러 책이름, 저자, 가격 출력하기

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
```

```python
url='https://book.naver.com/bestsell/bestseller_list.nhn'
```

```python
driver=webdriver.Chrome(ChromeDriverManager().install())
driver.implicitly_wait(30)
```

```python
driver.get(url)
obj=BeautifulSoup(driver.page_source, 'html.parser')
```

```python
obj
```

```python
book_page_urls=[]

for index in range(0,25):
    dl_data=obj.find('dt', {'id':'book_title_'+str(index)})
    link=dl_data.select('a')[0].get('href')
    book_page_urls.append(link)
    
for index, book_page_url in enumerate(book_page_urls):
    driver.get(book_page_url)
    obje=BeautifulSoup(driver.page_source, 'html.parser')
    
    title=obje.find('meta', {'property':'og:title'}).get('content')
    author=obje.find('dt',text='저자').find_next_siblings('dd')[0].text.strip()
    image=obje.find('meta', {'property':'og:image'}).get('content')
    url=obje.find('meta', {'property':'og:url'}).get('content')
    
    dd=obje.find('dt', text='가격').find_next_siblings('dd')[0]
    salePrice=dd.select('div.lowest strong')[0].text
    originalPrice=dd.select('div.lowest span.price')[0].text
    
    print(index+1, title, author, image, url, originalPrice, salePrice)
```

### 기상청의 동네예보 가져오기

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
html=urlopen('https://www.weather.go.kr/weather/forecast/timeseries.jsp')

soup=BeautifulSoup(html,'lxml')
soup
```

```python
table=soup.find('table', class_='forecastNew3')
```

```python
tr=table.tbody.tr
for t in tr.children:
    if t.name=='th':
        if t['scope']=='colgroup':
            num=int(t['colspan'])
            
            for i in range(num):
                print(t.get_text(), end=' ')
```
