# 3. 도시 분할 계획

```python
# 부호 하나가 ..ㅋㅋ

import heapq
def find_mom(moms,x):
    if moms[x]!=x:
        moms[x]=find_mom(moms, moms[x])
        
    return moms[x]

def coconut(graph, moms):
    res=[]

    for C,A,B in graph:
        
        a=find_mom(moms,A)
        b=find_mom(moms,B)
        print('C,A,B:',C,A,B)
        print(A,'의 부모',a)
        print(B,'의 부모',b)
        if a<b:
            moms[b]=a
            print(a,'의 부모',moms[a],'를 ',b,'에 부여',C)
        elif b<a:
            moms[a]=b
            print(b,'의 부모',moms[b],'를 ',a,'에 부여',C)
        else: # cycle
            print('싸이클이라 패스')
            continue # 여기 조건이 문제인가
        
        print(A,'to',B,':',C)
        res.append(C)
        print()
        
    return res

graph=[]

N,M=map(int, input().split())
for _ in range(M):
    A,B,C=map(int,input().split())
    #heapq.heappush(graph,(C,A,B))
    graph.append( (C,A,B) )

graph.sort()

#print('정렬 결과:',graph)
    
moms=[i for i in range(0,N+1)]

result=coconut(graph, moms)

print('result:',result)
del result[result.index(max(result))]

print(sum(result))
```

- 문제 해설

```python
def find_parent(parent,x):
    if parent[x]!=x:
        parent[x]=find_parent(parent, parent[x])
        
    return parent[x]

def union_parent(parent, a,b):
    a=find_parent(parent,a)
    b=find_parent(parent,b)
    
    if a<b:
        parent[b]=a
    else:
        parent[a]=b
        
v,e=map(int,input().split())
parent=[0]*(v+1)

edges=[]
result=0

for i in range(1, v+1):
    parent[i]=i
    
for _ in range(e):
    a,b,cost=map(int,input().split())
    edges.append( (cost,a,b) )
    
edges.sort()
last=0

for edge in edges:
    cost,a,b=edge
    
    if find_parent(parent,a)!=find_parent(parent,b):
        union_parent(parent,a,b)
        result+=cost
        last=cost
        
print(result-last)
```
