- DFS

```python
maze=[] # 지도

N,M=map(int,input().split()) # 행, 열

for _ in range(N):
    maze.append(list(map(int,input())))
    
count=1
    
def solution(graph,x,y):
    if graph[len(graph)-1][len(graph[0])-1]==0:
        print('목적지에 이미 도달함')
        return 0
    
    stack=[(x,y)]
    graph[x][y]=0 # 방문처리
    global count
    dx=[-1,1,0,0]
    dy=[0,0,-1,1] # 상하좌우
    
    while stack:
        if graph[len(graph)-1][len(graph[0])-1]==0:
                    print('도착해서 쫑낼게여')
                    return 0
                
        out=0
        nx, ny=stack.pop() # 탐색 시작
        print('***********탐색:',nx,ny,'************')
        
        if nx==(N-1) and ny==(M-1): # 도착?
            print('도착',nx,ny)
            break
            
        print('--------(',nx,ny,') 상하좌우 탐색---------')
        
        for i in range(4):
            if graph[len(graph)-1][len(graph[0])-1]==0:
                    print('도착해서 쫑낼게여')
                    return 0

            mx=nx+dx[i]
            my=ny+dy[i]
            
            print(nx,ny,'의 인접노드',mx,my)
            
            if mx<0 or mx>=N or my<0 or my>=M:
                print('틀 벗어남',mx,my)
                print()
                out+=1
                print('현재 out',out)
                if out==4:
                    print('막다름')
                    count-=1
                    break
                continue
                
            if graph[mx][my]==1:
                stack.append( (mx,my) ) # 삽입
                graph[mx][my]=0 # 방문처리
                print(mx,my,'방문처리 후 탐색 시작, 현재 방문 수는',count+1)
                count+=1
                solution(graph,mx,my)
            else: # 틀 안이지만 괴물
                print('틀 안이지만 괴물',mx,my)
                out+=1
                print('현재 out',out)
                if out==4:
                    print('막다름. 줄인 카운트는',count-1)
                    count-=1 # 방문 무효
                    break
                continue
        
        print(nx,ny,'에 대한 함수 종료-------------')
                
    return

solution(maze,0,0)
#print(count)
pd.DataFrame(maze)
```

- 답안 예시

```python
from collections import deque

n,m= map(int,input().split())
graph=[]

for i in range(n):
    graph.append(list(map(int,input())))
    
dx=[-1,1,0,0]
dy=[0,0,-1,1]

def bfs(x,y):
    queue=deque()
    queue.append((x,y))
    
    while queue:
        x,y=queue.popleft()
        
        for i in range(4):
            nx=x+dx[i]
            ny=y+dy[i]
            
            if nx<0 or ny<0 or nx>=n or ny>=m:
                continue
                
            if graph[nx][ny]==0:
                continue
                
            if graph[nx][ny]==1:
                graph[nx][ny]=graph[x][y]+1
                queue.append((nx,ny))
                
    return graph[n-1][m-1]

bfs(0,0)

pd.DataFrame(graph)
```
