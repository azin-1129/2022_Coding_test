# 3. 도시 분할 계획

```python
import heapq
def find_mom(moms,x):
    if moms[x]!=x:
        moms[x]=find_mom(moms, moms[x])
        
    return moms[x]

def coconut(graph, moms):
    res=[]

    for C,A,B in graph:
        
        a=find_mom(moms,A)
        b=find_mom(moms,B)
        print(A,'의 부모',a)
        print(B,'의 부모',b)
        if a<b:
            moms[b]=a
            print(a,'의 부모',moms[a],'를 ',b,'에 부여',C)
        elif b>a:
            moms[a]=b
            print(b,'의 부모',moms[b],'를 ',a,'에 부여',C)
        else: # cycle
            continue # 여기 조건이 문제인가
        
        print(A,'to',B,':',C)
        res.append(C)
        
    return res

graph=[]

N,M=map(int, input().split())
for _ in range(M):
    A,B,C=map(int,input().split())
    #heapq.heappush(graph,(C,A,B))
    graph.append( (C,A,B) )

graph.sort()

print('정렬 결과:',graph)
    
moms=[i for i in range(0,N+1)]

result=coconut(graph, moms)

print('result:',result)
del result[result.index(max(result))]

print(sum(result))
```
