```python
#223p
def f(n):
    global N,count
    
    if n==N:
        count+=1
        print("함수 종료")
        return 0
    
    print(n,"입력됨",n+1,n+2,n+2,"재귀중")
    return f(n+1),f(n+2),f(n+2)


N=int(input())

count=0

x,y,z=1,2,2

f(x)
f(y)
f(z)

print(count)
```
