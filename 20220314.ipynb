```python
# 217p 답안 예시

X=int(input())
d=[0]*1000001

for x in range(2,X+1):
    d[x]=d[x-1]+1
    print("d[x]=",d[x],"x=",x)
    
    if x%2==0:
        print("d[x//2]+1",d[x//2]+1)
        d[x]=min(d[x],d[x//2]+1)
        print("!d[x]:",d[x])
    if x%3==0:
        print("d[x//3]+1",d[x//3]+1)
        d[x]=min(d[x],d[x//3]+1)
        print("!d[x]:",d[x])
    if x%5==0:
        print("d[x//5]+1",d[x//5]+1)
        d[x]=min(d[x],d[x//5]+1)
        print("!d[x]:",d[x])
        
for a in range(26):
    print(d[a],end=" ")
    
print()
print(d[X])
```
