# 3. 도시 분할 계획

```python
# 부호 하나가 ..ㅋㅋ

import heapq
def find_mom(moms,x):
    if moms[x]!=x:
        moms[x]=find_mom(moms, moms[x])
        
    return moms[x]

def coconut(graph, moms):
    res=[]

    for C,A,B in graph:
        
        a=find_mom(moms,A)
        b=find_mom(moms,B)
        print('C,A,B:',C,A,B)
        print(A,'의 부모',a)
        print(B,'의 부모',b)
        if a<b:
            moms[b]=a
            print(a,'의 부모',moms[a],'를 ',b,'에 부여',C)
        elif b<a:
            moms[a]=b
            print(b,'의 부모',moms[b],'를 ',a,'에 부여',C)
        else: # cycle
            print('싸이클이라 패스')
            continue # 여기 조건이 문제인가
        
        print(A,'to',B,':',C)
        res.append(C)
        print()
        
    return res

graph=[]

N,M=map(int, input().split())
for _ in range(M):
    A,B,C=map(int,input().split())
    #heapq.heappush(graph,(C,A,B))
    graph.append( (C,A,B) )

graph.sort()

#print('정렬 결과:',graph)
    
moms=[i for i in range(0,N+1)]

result=coconut(graph, moms)

print('result:',result)
del result[result.index(max(result))]

print(sum(result))
```

- 문제 해설

```python
def find_parent(parent,x):
    if parent[x]!=x:
        parent[x]=find_parent(parent, parent[x])
        
    return parent[x]

def union_parent(parent, a,b):
    a=find_parent(parent,a)
    b=find_parent(parent,b)
    
    if a<b:
        parent[b]=a
    else:
        parent[a]=b
        
v,e=map(int,input().split())
parent=[0]*(v+1)

edges=[]
result=0

for i in range(1, v+1):
    parent[i]=i
    
for _ in range(e):
    a,b,cost=map(int,input().split())
    edges.append( (cost,a,b) )
    
edges.sort()
last=0

for edge in edges:
    cost,a,b=edge
    
    if find_parent(parent,a)!=find_parent(parent,b):
        union_parent(parent,a,b)
        result+=cost
        last=cost
        
print(result-last)
```

# 4. 커리큘럼

```python
from collections import deque

N=int(input()) # 강의 수
hour=[0]*(N+1) # 각 강의 수강 시간
pre_subject=[0]*(N+1) # 선수 과목 목록
visited=[False]*(N+1)
result=[0]*(N+1)

for i in range(1,N+1):
    hour_presub=list(map(int,input().split()))
    hour[i]=hour_presub[0] # 수강 시간
    pre_subject[i]=[item for item in hour_presub[1:len(hour_presub)] if item!= -1]
    
print('hour:',hour)
    
start=pre_subject.index([]) # 시작 과목번호
result[start]=hour[start] # 시작 과목 소요시간
visited[start]=True
visited[0]=True

print('start:',start,'time:',result[start])
print('pre_subject:',pre_subject)

def sorting(start):
    global pre_subject,hour,result,visited
    
    temp=[ [] for _ in range(N+1) ]
    
    q=deque([start]) # 큐 삽입. 1단계
    
    while q:
        pre_sub=q.popleft()
        
        for pre_list in pre_subject:
            print(pre_list,'안에',pre_sub,'찾는중')
            if visited[pre_subject.index(pre_list)]==True: # 이미 방문했다면
                continue
            
            if pre_sub in pre_list: # 혹시나 강의 내 선수과목 중 꺼낸 노드 있으면
                temp[pre_subject.index(pre_list)].append(hour[pre_sub]) # 선수과목 시간 리스트에 꺼낸 노드 시간 추가
                print('temp:',temp)
                del pre_list[pre_list.index(pre_sub)] # 제거
                print('pre_subject 갱신됨',pre_subject)
            
            if pre_list==[]: # 진입차수 0 되면 << 이거 [] 많아서 구분 못하는거같은데
                q.append(pre_subject.index(pre_list)) # 큐에 추가
                print('오류나는값:',pre_subject.index(pre_list))
                res[pre_subject.index(pre_list)]=max(temp[pre_subject.index(pre_list)])
                visited[pre_subject.index(pre_list)]=True
                
    return temp

for res in sorting(start):
    print(res)
```
