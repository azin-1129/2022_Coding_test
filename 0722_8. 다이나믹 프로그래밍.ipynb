### 1. 다이나믹 프로그래밍

- 중복되는 연산 줄이기

```python
def fibo(x):
    if x==1 or x==2:
        return 1
    return fibo(x-1)+fibo(x-2)

print(fibo(4))
```

```python
d=[0]*100 # 메모이제이션 리스트

def fibo(x):
    if x==1 or x==2:
        return 1
    
    if d[x]!=0:
        return d[x]
    
    d[x]=fibo(x-1)+fibo(x-2)
    
    return d[x]

print(fibo(99))
```

```python
d=[0]*100

def pibo(x):
    print('f('+str(x)+')',end=' ')
    if x==1 or x==2:
        return 1
    
    if d[x]!=0:
        return d[x]
    
    d[x]=pibo(x-1)+pibo(x-2)
    
    return d[x]

pibo(6)
```

### 실전문제

- 2. 1로 만들기

```python
def sibal(x, count):
    global DP
    #print('x',x,'count',count)
    #print(DP)
    if x==1:
        #print('=======이터레이션 끝===========')
        return 0
    
    
    
    if x%5==0:
        if DP[x//5]!=0:
            DP[x//5]=min(DP[x//5], count+1)
        else:
            DP[x//5]=(count+1)
        #print('**********',x,'//5','=',x//5,'**********')
        sibal(x//5, count+1)
    if x%3==0:
        if DP[x//3]!=0:
            DP[x//3]=min(DP[x//3], count+1)
        else:
            DP[x//3]=(count+1)
        #print('**********',x,'//3','=',x//3,'**********')
        sibal(x//3, count+1)
    if x%2==0:
        if DP[x//2]!=0:
            DP[x//2]=min(DP[x//2], count+1)
        else:
            DP[x//2]=(count+1)
        #print('**********',x,'//2','=',x//2,'**********')
        
        sibal(x//2, count+1)
    
    if DP[x-1]!=0:
        DP[x-1]=min(DP[x-1], count+1)
    else:
        DP[x-1]=(count+1)
    #print('**********',x,'-1','=',x-1,'**********')

    sibal(x-1, count+1)
    
X=int(input())
DP=[0]*(X+1)

sibal(X,0)

print(DP[1])
```

- 답안 예시

```python
x=int(input())

d=[0]*27

for i in range(2, x+1):
    d[i]=d[i-1]+1
    
    print(d)
    if i%2==0:
        d[i]=min(d[i], d[i//2]+1) # 나머지가 0 나오면 곤란해서 i가 2부터 시작하는 듯
    if i%3==0:
        d[i]=min(d[i], d[i//3]+1)
    if i%5==0:
        d[i]=min(d[i], d[i//5]+1) # 만드는 과정. 쌓아감
        
    # 1씩 더하는 거랑 min으로 (2,3,5) 더했을 때 어느게 최소인지 감별하고 d[i]에 대입
    
    print('--------갱신--------')
    print(d)
    print()
        
print(d[x])
```
